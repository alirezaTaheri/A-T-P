```python
final = 0
def checkMacd(macd,signal, maxShare):
    if macd[0] > signal[0]:
        return maxShare
    else:
        return 0
    
def checkRsi(rsi, maxShare):
    if rsi > 80:
        return 0
    if rsi < 20:
        return maxShare
    share = maxShare/60
    return maxShare - share*(rsi-20)

def checkEmaDiff(short_ema, long_ema):
    value = 0
    print("today Short EMA:",short_ema[0], "Long EMA:",long_ema[0], "Diff: ", short_ema[0] - long_ema[0])
    for i in range(0,-4, -1):
        if short_ema[i] - long_ema[i] < 0:
            print(i,"th Day Short EMA:",short_ema[i], "Long EMA:",long_ema[i],"Diff",short_ema[i] - long_ema[i])
            return value
        elif short_ema[i] - long_ema[i] > short_ema[i-1] - long_ema[i-1]:
            print(i-1,"th Day Short EMA:",short_ema[i-1], "Long EMA:",long_ema[i-1],"Diff",short_ema[i-1] - long_ema[i-1])
            value += 2.5
        else:
            print(i-1,"th Day Short EMA:",short_ema[i-1], "Long EMA:",long_ema[i-1] ,"Diff",short_ema[i-1] - long_ema[i-1])
            return value
    return value

def checkBollinger(close, bollinger, maxShare):
    top = bollinger.lines.top[0]
    bottom = bollinger.lines.bot[0]
    if close[0] > top or close[0] < bottom:
        return 0
    share = maxShare/(top - bottom)
    return maxShare - share*(close[0] - bottom)

def checkStochastic(stoch, maxShare_comp, maxShare_relative):
    totalValue = 0
    if stoch.fastk[0] > stoch.fastd[0]:
        totalValue += maxShare_comp
    print("@STOCH, stoch comparison:", totalValue)
    if stoch.fastk[0] > 85:
        return totalValue
    if stoch.fastk[0] < 15:
        totalValue += maxShare_relative
        return totalValue
    share = maxShare_relative/70
    totalValue += maxShare_relative - share*(stoch.fastk[0] - 15)
    return totalValue
 
def checkDate(fromdate, currentdate):
    if currentdate < fromdate + pd.DateOffset(years=2):
        return False
    return True

def sliceDate(data,currentDay,freq, period):
    if 'm' in period:
        return data[currentDay-pd.DateOffset(months=freq):currentDay]
    if 'y' in period:
        return data[currentDay-pd.DateOffset(years=freq):currentDay]
    
def checkCF(close, day, firstMaxShare, secondMaxShare, thirdMaxShare):
    close.index = pd.to_datetime(close.index, format='%Y-%m-%d')
    diff_6_first = sliceDate(close, day,6, 'm')
    diff_6_second = sliceDate(close, day,1, 'y')[:sliceDate(close, day,6,'m' ).first_valid_index()-pd.DateOffset(days=1)]
    diff_1_year = sliceDate(close, day,2, 'y')[:sliceDate(close, day,1,'y' ).first_valid_index()-pd.DateOffset(days=1)]
    diff_6_first_min = diff_6_first.min()
    diff_6_first_max = diff_6_first.max()
    diff_6_second_min = diff_6_second.min()
    diff_6_second_max = diff_6_second.max()
    diff_1_year_min = diff_1_year.min()
    diff_1_year_max = diff_1_year.max()
    exist = [True , False, False]
    if close[day] <= diff_6_second_max and close[day] >= diff_6_second_min:
        exist[1] = True
    if close[day] <= diff_1_year_max and close[day] >= diff_1_year_min:
        exist[2] = True
    if not exist[2] and not exist[1]:
        firstMaxShare = 20
        secondMaxShare = 0
        thirdMaxShare = 0
    if not exist[2] and exist[1]:
        firstMaxShare = 10
        secondMaxShare = 10
        thirdMaxShare = 0
    if not exist[1] and exist[2]:
        firstMaxShare = 10
        secondMaxShare = 0
        thirdMaxShare = 10
    print("Current Close: ", close[day])
    print(diff_6_first.index.min(), diff_6_first.index.max())
    print('Check between', diff_6_first_min,'  ', diff_6_first_max, exist[0])
    print(diff_6_second.index.min(), diff_6_second.index.max())
    print('Check between', diff_6_second_min,'  ', diff_6_second_max, exist[1])
    print(diff_1_year.index.min(), diff_1_year.index.max())
    print('Check between', diff_1_year_min,'  ', diff_1_year_max, exist[2])
    totalValue = 0
    print('Shares: ', firstMaxShare, secondMaxShare, thirdMaxShare)
    if close.loc[day] >= diff_6_first_min and close.loc[day]<= diff_6_first_max:
        share = firstMaxShare/ (diff_6_first_max - diff_6_first_min)
        totalValue += firstMaxShare - share*(close[day] - diff_6_first_min)
    if close.loc[day] >= diff_6_second_min and close.loc[day] <= diff_6_second_max:
        share = secondMaxShare/ (diff_6_second_max - diff_6_second_min)
        totalValue += secondMaxShare - share*(close[day]-diff_6_second_min)
    if close[day] >= diff_1_year_min and close.loc[day] <= diff_1_year_max:
        share = thirdMaxShare/ (diff_1_year_max - diff_1_year_min)
        totalValue += thirdMaxShare - share*(close[day]-diff_1_year_min)
    return totalValue

def checkMfi(mfi, maxShare):
    if mfi[0] > 90:
        return 0
    if mfi[0] < 10:
        return maxShare
    share = maxShare/80
    return maxShare - share*(mfi[0] - 10)

def checkAll(mfi, rsi, bollinger, close, stoch, date):
    algoShares = {'mfi':20, 'rsi':20, 'bollinger':20, 'stochastic_comp':10, 'stochastic_relative':10, 'cf':20}
    rsiValue = checkRsi(rsi, algoShares['rsi'])
    mfiValue = checkMfi(mfi, algoShares['mfi'])
    bollingerValue = checkBollinger(close, bollinger, algoShares['bollinger'])
    stochasticValue = checkStochastic(stoch, algoShares['stochastic_comp'], algoShares['stochastic_relative'])
    cfValue = checkCF(indexedClose, pd.Timestamp(date), 8, 7, 5)    
    print('////////////////',date,'/////////////////////')
    print("RSI", rsiValue)
    print("MFI",mfiValue)
    print("Bollinger Bands", bollingerValue)
    print("Stochastic Value", stochasticValue)
    print("CF Value", cfValue)
    results = rsiValue + mfiValue + bollingerValue + stochasticValue + cfValue
    print("##################################### Results = ", results,"%")
    return results
```
